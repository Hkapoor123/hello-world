3. Making the data readable involves cleaning and organizing it in a way that allows easy comprehension. The goal is to remove any unnecessary clutter and retain only the relevant information needed for analysis.


Main Steps Taken for Data Wrangling:

```python
#import pandas, io, seaborn and numpy libraries 

import pandas as pd
import io
import numpy as np
import seaborn as sns
```

```python
##Data Wrangling: Getting all of the data in, and in the desired formats!

# World Energy Comsumption dataset
WEC_Data = pd.read_csv('Datasets/World Energy Consumption.csv')


# Billionaire Index
#Reading the CSV files with dataset showing first 5 and last 5 columns
Forbes_20 = pd.read_csv('Datasets/Forbes_Billionaires_20.csv')
Forbes_21 = pd.read_csv('Datasets/Forbes_Billionaires_21.csv')
Forbes_22 = pd.read_csv('Datasets/Forbes_Billionaires_22.csv', index_col=0)
Forbes_23 = pd.read_csv('Datasets/Forbes_Billionaires_23.csv', encoding='latin-1')


#Understanding the basica datasets and shapes of WEC
print("World Energy Consumption")
print(WEC_Data.shape)
print(WEC_Data.describe)




```

```python
#Understanding the basica datasets and shapes of Forbes datasets 
print("Forbes billionares 2020")
print(Forbes_20.shape)
print(Forbes_20.describe)

print("Forbes billionares 2021")
print(Forbes_21.shape)
print(Forbes_21.describe)

print("Forbes billionares 2022")
print(Forbes_22.shape)
print(Forbes_22.describe)

print("Forbes billionares 2022")
print(Forbes_23.shape)
print(Forbes_23.describe)
```

```python
#Adding a columns for the year to help in differntiating after concatatanation 
Forbes_20['Year'] = '2020'
Forbes_21['Year'] = '2021'
Forbes_22['Year'] = '2022'
Forbes_23['Year'] = '2023'

#Renaming columns to match 2020 data format 
Forbes_21 = Forbes_21.rename(columns={'Industry': 'Industries',
                                      'NetWorth': 'Net Worth',
                                      'Country': 'Country/Territory'})
Forbes_22 = Forbes_22.rename(columns={'industry': 'Industries',
                                      'networth': 'Net Worth',
                                      'country': 'Country/Territory',
                                      'rank': 'Rank', 
                                      'name': 'Name',
                                      'source': 'Source',
                                      'age': 'Age'})
Forbes_23 = Forbes_23.rename(columns={'Industry': 'Industries',
                                      'source': 'Source',
                                      'Country': 'Country/Territory'})

#Combining data frames 2020, 2021, 2022 into "combined_billionaire"
combined_billionaire= pd.concat([Forbes_20, Forbes_21, Forbes_22, Forbes_23])

combined_billionaire
```

# Data Cleaning

Describe your data cleaning steps here.

```python
##     Data Cleaning

#Drops column where the Name is NaN
combined_billionaire = combined_billionaire.dropna(subset=['Name'])

#Replacing NaN values in Age with -1 for int casting
combined_billionaire['Age'] = combined_billionaire['Age'].fillna(-1)

#Converting numeric values to ints/floats for later data analysis
combined_billionaire["Net Worth"] = combined_billionaire["Net Worth"].replace("[$,B]", "", regex=True).astype(float)
combined_billionaire = combined_billionaire.astype({"Year":int,
                                                    "Age":int,
                                                   "Rank":int})

#Renaming columns for accuracy after changes if needed
combined_billionaire = combined_billionaire.rename(columns={'Net Worth' : 'Net Worth (Billions)'})

#Renaming names for consistency
combined_billionaire['Name'] = combined_billionaire['Name'].str.title()

#Removing white space from Industries for consistency
combined_billionaire['Industries'] = combined_billionaire['Industries'].str.strip()

combined_billionaire
```

```python
#Seaborn graph showing tech industry billionaires over time
#Note names removed, because a list of 100 names is not helpful for anyone
combined_billionaire = combined_billionaire.sort_values("Net Worth (Billions)", ascending=False, ignore_index=True)

reduced_dataset = combined_billionaire.query("Industries == 'Technology'")

YearxWorth = sns.lineplot(data=reduced_dataset, x='Year', y='Net Worth (Billions)', hue="Name", legend=False)
```

```python
#Seaborn graph showing the average net worth of industry billionaires over time
IndustriesGroup = combined_billionaire.groupby(['Industries', 'Year'])

industries_mean = IndustriesGroup.mean()

YearxWorth = sns.lineplot(data=industries_mean, x='Year', y='Net Worth (Billions)', hue="Industries")

# Put the legend out of the figure
YearxWorth.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```

# Data Cleaning


1. **Handling NaN Values and Linear Connections:**
   - We started by cleaning the dataset, particularly focusing on columns with `np.float64` data type.
   - The `linearize_nan` function was applied to fill or linearize the NaN values in these columns, creating a smoother trend for missing data points.
   - This linearization process helped maintain a logical progression in the data, especially useful for time series or continuous data where gradual changes were expected.

2. **Rounding Floats to Two Digits:**
   - After handling NaN values, we rounded all floating-point numbers in the dataset to two decimal places.
   - Rounding ensured a cleaner representation of numerical values without excessive precision, making it more suitable for presentation and analysis.

3. **Selecting Renewable Energy Columns:**
   - We isolated specific columns related to renewable energy statistics, including country, year, and various renewable energy consumption metrics.
   - This selection aimed to focus on relevant data related to renewable energy trends and consumption.

4. **Selecting Fossil Fuels Columns:**
   - Similarly, we isolated columns related to fossil fuel statistics, including country, year, and various fossil fuel consumption metrics.
   - This enabled a focused analysis of fossil fuel consumption trends and energy share in the dataset.

## Achievements:
- The data cleaning process improved data integrity and consistency by handling missing values in a logical and gradual manner.
- Linearizing NaN values helped fill gaps in the data, providing a more continuous representation of energy consumption trends.
- Rounding the float values to two decimal places reduced unnecessary precision, making the data more concise and presentable.
- By isolating and organizing the renewable energy and fossil fuel columns separately, we facilitated targeted analysis and insights into each energy sector.
- Overall, this data cleaning and organization process prepared the dataset for further analysis and visualization, enabling a clearer understanding of renewable energy and fossil fuel consumption patterns over time for different countries.

````python
# Creating helper functions to handle NaN values

def linearize_nan_end(col, nulCheck):
    return_col = col
    idx_1 = -1
    while nulCheck.iloc[idx_1]:
            idx_1 -= 1
    mean = (col[idx_1-1] + col[idx_1])/2
    while idx_1 < -1:
        idx_1 += 1
        return_col[idx_1] = return_col[idx_1-1] + mean
    return return_col

def linearize_nan(col):
    return_col = [0 for n in range(len(col))]
    nulCheck = col.isnull()
    toNanlst = []
    idx_1 = 0
    while idx_1 < len(col):
        # If you encounter a NaN
        if nulCheck[idx_1] == True:
            # If the first value is NaN, treat as 0
            if idx_1 == 0:
                return_col[idx_1] = 0
                idx_1 += 1
                break
            elif idx_1 == len(col)-1:
                # No valid end point! Change tactics!
                return linearize_nan_end(return_col, nulCheck)
            else:
                # Append the previous value to toNanlst, and a 0 for each NaN until the next valid number
                toNanlst.append(col[idx_1-1])
                idx_2 = idx_1
                while idx_2 < len(col):
                    if idx_2 == len(col)-1:
                        # No valid end point! Change tactics!
                        return linearize_nan_end(return_col, nulCheck)
                    elif nulCheck[idx_2] == True:
                        toNanlst.append(0)
                    else:
                        toNanlst.append(col[idx_2])
                        break
                    idx_2 += 1
                nanMean = (toNanlst[-1]-toNanlst[0])/(len(toNanlst)-1)
                toNanlst = []
                while idx_1 < idx_2:
                    return_col[idx_1] = round(return_col[idx_1-1]+nanMean, 3)
                    idx_1 += 1
        return_col[idx_1] = col[idx_1]
        idx_1 += 1
    return return_col
````

````python
# Creating linear connections between NaN data.
for col in merged_wec_df.columns:
    if merged_wec_df[col].dtype == np.float64:
        merged_wec_df[col] = linearize_nan(merged_wec_df[col])

# Rounding all floats to 2 digits
for col in merged_wec_df.columns:
    if merged_wec_df[col].dtype == np.float64:
        merged_wec_df[col] = merged_wec_df[col].round(2)
        
# only using colums that say renewable energy and those that say fossil fuels 
renewable = ['Country', 'Year',
'Renewable Consumption Change (%)',
'Renewable Energy Share (%)',
'Renewable Consumption Change (TWh)',
'Renewable Consumption (TWh)',
'Renewable Energy Per Capita (KWh)']
renewable_df = merged_wec_df[renewable].copy()
renewable_df.reset_index(drop=True, inplace=True)

nrenewable = ['Country', 'Year',
'Fossil Consumption Change (%)',
'Fossil Energy Share (%)',
'Fossil Consumption Change (TWh)',
'Fossil Consumption (TWh)',
'Fossil Energy Per Capita (KWh)']
nrenewable_df = merged_wec_df[nrenewable].copy()
nrenewable_df.reset_index(drop=True, inplace=True)
````

````python
formula = 'Q("Net Worth ($US billion)") ~ Q("Renewable Consumption Change (%)")'

# Create the design matrices using Patsy
outcome_1, predictors_1 = patsy.dmatrices(formula, data=all_data, return_type='dataframe')

# Perform linear regression using matrix multiplication
coefficients = np.linalg.inv(predictors_1.T @ predictors_1) @ predictors_1.T @ outcome_1

# Get the intercept and coefficient values
intercept, coefficient = coefficients.values.flatten()

print(f"Intercept: {intercept}")
print(f"Coefficient: {coefficient}")
````

Intercept: 655.65
Coefficient: -12.36

The linear regression model can be represented as follows:

Net Worth ($US billion) = 655.65 - 12.36 * Renewable Consumption Change (%)

This means that, according to the linear regression model, there is a negative relationship between the net worth of a country in billions of US dollars and its renewable consumption change percentage. Specifically, for every one percentage point increase in renewable consumption change, the net worth of the country is expected to decrease by approximately 12.36 billion US dollars. It's important to remember that this is a statistical model, and causality or other factors may not be fully accounted for in this simple linear regression. Additionally, it's essential to assess the model's accuracy and consider other variables or models for a comprehensive analysis.
